### 1. Principal Component Analysis (PCA)

1. The main goal of PCA is to find a new set of variables, called principal components, that are linear combinations of the original variables. These principal components are chosen in such a way that they capture the maximum amount of variation in the data. The first principal component explains the most variation in the data, the second principal component explains the second most variation, and so on.

2. PCA works by calculating the eigenvectors and eigenvalues of the covariance matrix of the data. The eigenvectors represent the directions in which the data varies the most, and the eigenvalues represent the amount of variation along each eigenvector. By keeping only a subset of the principal components with the highest eigenvalues, we can reduce the dimensionality of the data.

3. PCA is often used as a pre-processing step in machine learning tasks to reduce the number of features and remove redundant information. It can also be used for data visualization, as it allows us to represent high-dimensional data in a lower-dimensional space, typically 2D or 3D, which can be easily plotted and analyzed.

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
data = load_iris()
X = data.data
y = data.target
```

```python
# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
# Plot the reduced data
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
plt.title("PCA")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

### 2.t-SNE

1. t-SNE (t-Distributed Stochastic Neighbor Embedding) is a popular dimensionality reduction technique used for visualizing high-dimensional data in a lower-dimensional space, typically 2D or 3D. It is particularly effective in capturing complex structures and relationships in the data.

2. The main idea behind t-SNE is to map each data point from the high-dimensional space to a point in the low-dimensional space while preserving the relationships and similarities between the data points as much as possible.

3. Here is an overview of how t-SNE works:

* Computing Pairwise Similarities: First, t-SNE computes the pairwise similarities between all data points in the high-dimensional space. Similarities are typically calculated using a Gaussian kernel or a distance metric such as Euclidean distance.

* Constructing Conditional Probabilities: t-SNE then converts the pairwise similarities into conditional probabilities that represent the likelihood of selecting a neighbor data point given another data point. The probabilities are calculated by normalizing the pairwise similarities.

* Creating Low-Dimensional Embedding: Next, t-SNE constructs a similar probability distribution in the low-dimensional space, aiming to preserve the relationships captured in the high-dimensional space. It uses gradient descent to minimize the divergence between the high-dimensional and low-dimensional distributions.

* Optimizing Embedding: During the optimization process, t-SNE iteratively moves the data points in the low-dimensional space to find an embedding that best represents the relationships between the data points. It uses a cost function that measures the similarity between the high-dimensional and low-dimensional probability distributions.

* Visualizing the Embedding: Finally, t-SNE maps the high-dimensional data points onto a two-dimensional or three-dimensional space, where each point represents an instance in the dataset. The visualization allows for the examination of the underlying structure and clusters within the data.

```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
data = load_iris()
X = data.data
y = data.target
```

```python
# Apply t-SNE
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X)
```

```python
# Plot the reduced data
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y)
plt.title("t-SNE")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.show()
```
