#### 1. Iris Dataset: This dataset contains measurements of different features of iris flowers, such as sepal length, sepal width, petal length, and petal width.

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

# Print the reduced dimensions
print(X_reduced.shape)
```

```python
type(y)
```

```python
import matplotlib.pyplot as plt
# Plot the reduced data
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y)
plt.title("PCA")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

#### 2. MNIST Dataset: This dataset consists of a large number of handwritten digits from 0 to 9. Each digit is represented as a 28x28 grayscale image. It is often used for image classification tasks, but can also be used for dimensionality reduction.

```python
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA

# Load the MNIST dataset
mnist = fetch_openml('mnist_784')
X = mnist.data
y = mnist.target

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

# Print the reduced dimensions
print(X_reduced.shape)
```

```python
import matplotlib.pyplot as plt
# Plot the reduced data
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y.astype('int'))
plt.title("PCA")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```
