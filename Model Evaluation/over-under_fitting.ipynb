```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Generate some random data
np.random.seed(0)
X = np.linspace(-3, 3, 100)
y_true = np.sin(X) + np.random.normal(0, 0.1, 100)
```

### Underfitting

1. **High bias**: The model is too simple to capture the complexity of the data, leading to a poor fit to both the training and testing data.
2. **Insufficient model capacity**: The model does not have enough parameters or complexity to capture the underlying patterns in the data.
3. **Underparameterization**: The model lacks the necessary flexibility to represent the relationships in the data accurately.
4. **Undergeneralization**: The model fails to capture important features or trends in the data, resulting in a lack of accuracy and predictive power.

#### degree = 2

```python
# Add polynomial features
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X.reshape(-1, 1))

# Split the data into training and testing sets
X_train, y_train = X_poly[:80], y_true[:80]
X_test, y_test = X_poly[80:], y_true[80:]

# Fit a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Evaluate the model on training and testing data
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Calculate mean squared errors
train_error = mean_squared_error(y_train, y_train_pred)
test_error = mean_squared_error(y_test, y_test_pred)

print(f'Training Error: {train_error:.3f}, Testing Error: {test_error:.3f}')
```

```python
# Plot the data and the model predictions
plt.scatter(X, y_true, label='True Data')
plt.plot(X, model.predict(X_poly), color='red', label='Model Prediction')
plt.ylim((-2,2))
plt.xlim((-3,3))
plt.title(f'Training Error: {train_error:.2f}, Testing Error: {test_error:.2f}')
plt.legend()
plt.show()
```

### Overfitting

1. **High variance**: The model is too complex and captures noise or random fluctuations in the training data, leading to poor generalization on unseen data.
2. **Overparameterization**: The model has more parameters than necessary to represent the underlying patterns in the data, making it prone to fitting the noise.
3. **Data memorization**: The model essentially memorizes the training data instead of learning the underlying relationships, resulting in poor performance on new data.
4. **Overly flexible**: The model can adapt too much to the training data, resulting in sensitivity to small changes in the input.

#### degree = 20

```python
# Add polynomial features
poly = PolynomialFeatures(degree=20)
X_poly = poly.fit_transform(X.reshape(-1, 1))

# Split the data into training and testing sets
X_train, y_train = X_poly[:80], y_true[:80]
X_test, y_test = X_poly[80:], y_true[80:]

# Fit a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Evaluate the model on training and testing data
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Calculate mean squared errors
train_error = mean_squared_error(y_train, y_train_pred)
test_error = mean_squared_error(y_test, y_test_pred)

print(f'Training Error: {train_error:.3f}, Testing Error: {test_error:.3f}')
```

```python
# Plot the data and the model predictions
plt.scatter(X, y_true, label='True Data')
plt.plot(X, model.predict(X_poly), color='red', label='Model Prediction')
plt.ylim((-2,2))
plt.xlim((-3,3))
plt.title(f'Training Error: {train_error:.4f}, Testing Error: {test_error:.4f}')
plt.legend()
plt.show()
```

### Balanced fit

1. **Appropriate model complexity**: The model strikes a balance between capturing the underlying patterns in the data without overfitting or underfitting.
2. **Optimal trade-off**: The model achieves a good balance between bias and variance, leading to reasonable generalization performance on unseen data.
3. **Good generalization**: The model accurately represents the underlying relationships in the data, allowing for reliable predictions on new, unseen instances.
4. **Properly calibrated**: The model is well-tuned to the data, accounting for both complexity and simplicity in a balanced manner.

#### degree = 3

```python
# Add polynomial features
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X.reshape(-1, 1))

# Split the data into training and testing sets
X_train, y_train = X_poly[:80], y_true[:80]
X_test, y_test = X_poly[80:], y_true[80:]

# Fit a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Evaluate the model on training and testing data
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Calculate mean squared errors
train_error = mean_squared_error(y_train, y_train_pred)
test_error = mean_squared_error(y_test, y_test_pred)

print(f'Training Error: {train_error:.3f}, Testing Error: {test_error:.3f}')
```

```python
# Plot the data and the model predictions
plt.scatter(X, y_true, label='True Data')
plt.plot(X, model.predict(X_poly), color='red', label='Model Prediction')
plt.ylim((-2,2))
plt.xlim((-3,3))
plt.title(f'Training Error: {train_error:.4f}, Testing Error: {test_error:.4f}')
plt.legend()
plt.show()
```
